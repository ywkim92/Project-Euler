```python
cipher = []
with open('p059.txt','r') as file:
    cipher = file.read().strip().split(',')
    cipher = list(int(a) for a in cipher)
```

```python
from IPython.display import Image
Image(filename='ascii_codes.png', width = 1000) 
```

```python
# checking whether a deciphered letter belongs to commonly used characters
# types of input and output are ascii_code numbers
def check(enciphered, key):
    plain = enciphered ^ key
    if plain in range(32, 94):
        return True
    elif plain in range(97,123):
        return True
    return False
```

```python
first = []
for small in range(97,123):
    count = 0
    for c in range(0,len(cipher),3):
        if not check(cipher[c], small):
            count +=1
            break
    if count==0:
        first.append(small)
first
```

```python
second = []
for small in range(97,123):
    count = 0
    for c in range(1,len(cipher),3):
        if not check(cipher[c], small):
            count +=1
            break
    if count==0:
        second.append(small)
second
```

```python
third = []
for small in range(97,123):
    count = 0
    for c in range(2,len(cipher),3):
        if not check(cipher[c], small):
            count +=1
            break
    if count==0:
        third.append(small)
third
```

```python
key = first+second+third
key_str = ''.join(list(chr(i) for i in key))
print(key, key_str)
```

```python
deciphered = cipher.copy()
for idx in range(len(cipher)):
    deciphered[idx] ^= key[idx%3]
```

```python
message = ''
for letter in deciphered:
    message += chr(letter)
print(message)
```
